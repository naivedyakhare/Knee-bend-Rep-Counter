```python
import mediapipe as mp
import cv2
import numpy as np
import time
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
```

```python
def calculateAngle(a,b,c):
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)
    
    radian = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radian*180.0/np.pi)
    
    if angle > 180:
        angle = 360 - 180
    return angle
```

```python
cap = cv2.VideoCapture("data/KneeBendVideo.mp4")
# cap = cv2.VideoCapture(0)
closer_knee = None
closer_heel = None
closer_hip = None

stage = None
startTime = None
endTime = None
curTime = 0
angle = None
prevAngle = None
POSX, POSY = [None, None]
FONT_SIZE = 1
totalRep = 0
with mp_pose.Pose(min_detection_confidence=0.5,min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        
        #Detect stuff and render
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        
        results = pose.process(image)
        
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        #EXTRACT LANDMARKS
        try:
            landmarks = results.pose_landmarks.landmark
            if landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].z < landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].z and closer_knee == None:
                closer_knee = "LEFT_KNEE"
                closer_heel = "LEFT_HEEL"
                closer_hip = "LEFT_HIP"
            elif closer_knee == None:
                closer_knee = "RIGHT_KNEE"
                closer_heel = "RIGHT_HEEL"
                closer_hip = "RIGHT_HIP"
            
            knee = [landmarks[mp_pose.PoseLandmark[closer_knee].value].x, landmarks[mp_pose.PoseLandmark[closer_knee].value].y]
            heel = [landmarks[mp_pose.PoseLandmark[closer_heel].value].x, landmarks[mp_pose.PoseLandmark[closer_heel].value].y]
            hip =  [landmarks[mp_pose.PoseLandmark[closer_hip].value].x, landmarks[mp_pose.PoseLandmark[closer_hip].value].y]
            
            prevAngle = angle
            angle = calculateAngle(heel, knee, hip)
            
            if prevAngle - angle > 20 or prevAngle - angle < -20:
                pass
            
            if angle > 160:
                stage = "down"
                endTime = time.time()
                if endTime - startTime < 8:
                    string = "You need to hold the pose for more than 8 seconds"
                    POSX, POSY = [100, 100]
                    FONT_SIZE = 0.7
                
            if angle < 140 and stage == "down":
                stage = "up"
                startTime = time.time()
                string = f"Timer: {str(int(curTime))}"
                POSX, POSY = [654, 100]
                totalRep += 1
                print(totalRep)
                
            if startTime == 0 and stage == "up":
                curTime = 0
                string = f"Timer: {str(int(curTime))}"
                POSX, POSY = [654, 100]
                totalRep += 1
                print(totalRep)

            elif stage == "up":
                curTime = time.time() - startTime
                string = f"Timer: {str(int(curTime))}"
                POSX, POSY = [654, 100]
        except:
            pass
        
        cv2.putText(image, string, 
                       (POSX, POSY),
                             cv2.FONT_HERSHEY_SIMPLEX, FONT_SIZE, (255,0,0), 2, cv2.LINE_AA
                       )
        
        #Render Detections
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS, 
                                  mp_drawing.DrawingSpec(color=(245, 117, 66), thickness=2, circle_radius=2),
                                  mp_drawing.DrawingSpec(color=(245, 66, 280), thickness=2, circle_radius=2)
                                 )
            
        cv2.imshow("Mediapipe Feed", image)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    cap.release()
    cv2.destroyAllWindows()
    
```
